```python
import cv2
from google.colab.patches import cv2_imshow
from PIL import Image
```

##ROC Curves

```python
rocA = cv2.imread('/content/drive/MyDrive/TQP-AutoScore/Figure 2A.jpg')
rocB = cv2.imread('/content/drive/MyDrive/TQP-AutoScore/Figure 2B.jpg')
rocC = cv2.imread('/content/drive/MyDrive/TQP-AutoScore/Figure 2C.jpg')
rocD = cv2.imread('/content/drive/MyDrive/TQP-AutoScore/Figure 2D.jpg')
rocE = cv2.imread('/content/drive/MyDrive/TQP-AutoScore/Figure 2E.jpg')

roc1 = cv2.hconcat([rocA, rocB, rocC])
roc2 = cv2.hconcat([rocD, rocE])
cv2.imwrite('/content/drive/MyDrive/TQP-AutoScore/roc1.jpg', roc1, [cv2.IMWRITE_JPEG_QUALITY, 100])
cv2.imwrite('/content/drive/MyDrive/TQP-AutoScore/roc2.jpg', roc2, [cv2.IMWRITE_JPEG_QUALITY, 100])

roc1 = Image.open('/content/drive/MyDrive/TQP-AutoScore/roc1.jpg')
new_size = roc1.size
roc2 = Image.open('/content/drive/MyDrive/TQP-AutoScore/roc2.jpg')
old_size = roc2.size

new_im = Image.new("RGB", new_size, "White")
box = tuple((n - o) // 2 for n, o in zip(new_size, old_size))
new_im.paste(roc2, box)
new_im = new_im.save('/content/drive/MyDrive/TQP-AutoScore/roc2.jpg')

roc1 = cv2.imread('/content/drive/MyDrive/TQP-AutoScore/roc1.jpg')
roc2 = cv2.imread('/content/drive/MyDrive/TQP-AutoScore/roc2.jpg')

roc = cv2.vconcat([roc1, roc2])

cv2_imshow(roc)
cv2.imwrite('/content/drive/MyDrive/TQP-AutoScore/Figure 1.jpg', roc, [cv2.IMWRITE_JPEG_QUALITY, 100])
```

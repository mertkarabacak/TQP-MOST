```python
#Change variable names to field names.

data_dictionary = pd.read_csv("/content/drive/MyDrive/TQP-AutoScore/Modified Data Dictionary (TQP-AutoScore).csv", encoding = 'latin1', index_col = None, low_memory = False)
FieldNames = dict(zip(data_dictionary['Variable'], data_dictionary['Field Name']))
data.columns = data.columns.map(FieldNames)
```

```python
#Drop patients with unknown outcome of interest and discharges to hospice care.

survival_data = data.copy()

before = survival_data.shape[0]
survival_data = survival_data[survival_data['Discharge Disposition'].notna()]
survival_data = survival_data[survival_data['Total Length of Stay (Hours)'].notna()]
survival_data = survival_data[survival_data['Discharge Disposition'] != 'Discharged/Transferred to hospice care']

after = survival_data.shape[0]

excluded = before - after
print('Number of patients before exclusion: ', before)
print('Number of patients after exclusion: ', after)
print('Number of patients excluded with this criteria: ', excluded)
```

```python
#Define outcome of interest.

survival_data.loc[survival_data['Discharge Disposition'] == 'Deceased/Expired', 'label_status'] = 1
survival_data.loc[survival_data['label_status'] != 1, 'label_status'] = 0

survival_data['label_time'] = survival_data['Total Length of Stay (Hours)']

survival_data['label_status'].value_counts(normalize=False, dropna=False)
```

```python
#Drop outcome columns

outcomes = ['Total Length of Stay (Days)', 'Total Length of Stay (Hours)', 'Discharge Disposition', 'Hospital Complication - Cardiac Arrest with CPR', 'Hospital Complication - Catheter-Associated Urinary Tract Infection (CAUTI)', 'Hospital Complication - Central Line-Associated Bloodstream Infection (CLABSI)', 'Hospital Complication - Catheter-Related Blood Stream Infection', 'Hospital Complication - Deep Surgical Site Infection', 'Hospital Complication - Delirium', 'Hospital Complication - Alcohol Withdrawal Syndrome', 'Hospital Complication - Deep Vein Thrombosis (DVT)', 'Hospital Complication - Pulmonary Embolism', 'Hospital Complication - Extremity Compartment Syndrome', 'Hospital Complication - Graft or Prosthesis or Flap Failure', 'Hospital Complication - Unplanned Intubation', 'Hospital Complication - Acute Kidney Injury', 'Hospital Complication - Myocardial Infarction', 'Hospital Complication - Organ or Space Surgical Site Infection', 'Hospital Complication - Osteomyelitis', 'Hospital Complication - Pneumonia', 'Hospital Complication - Pressure or Decubitis Ulcer', 'Hospital Complication - Acute Respiratory Distress Syndrome (ARDS)', 'Hospital Complication - Unplanned Return to the OR', 'Hospital Complication - Severe Sepsis', 'Hospital Complication - Stroke  or  CVA', 'Hospital Complication - Superficial Surgical Site Infection', 'Hospital Complication - Unplanned Admission to the ICU', 'Hospital Complication - Urinary Tract Infection', 'Hospital Complication - Ventilator-Associated Pneumonia (VAP)', 'Total ICU Length of Stay']

survival_data = survival_data.drop(outcomes, axis = 1)
```
